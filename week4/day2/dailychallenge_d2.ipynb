Visualize the Decision Boundary

La fonction plot_decision_boundary sert à visualiser la "frontière de décision" de ton modèle. C'est une ligne (ou courbe) qui sépare les différentes classes de données, et cela nous aide à mieux comprendre comment le modèle prend ses décisions.
C'est très utile pour comprendre les performances de ton modèle et pour vérifier si la frontière de décision est raisonnable.

```python
# Fonction pour visualiser la frontière de décision
def plot_decision_boundary(model, X, y):
    # Crée une grille de points couvrant tout l'espace des données
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), 
                         np.arange(y_min, y_max, 0.02))
    
    # Prédictions sur toute la grille
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)  # Redimensionne pour correspondre à la grille

    # Trace la frontière de décision en coloriant les zones
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    
    # Trace les points de données réels
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')
    plt.title("Frontière de Décision")
    plt.show()

# Exemple d'utilisation
# Supposons que 'model' est déjà entraîné et que 'X' et 'y' sont tes données.
# plot_decision_boundary(model, X, y)
```
