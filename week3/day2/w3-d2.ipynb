```python
import numpy as np
import pandas as pd
```

```python

np.random.seed(0)  # For reproducibility
temperatures = np.random.uniform(-5, 35, (10, 12))  # 10 cities, 12 months
cities = ["City " + str(i) for i in range(1, 11)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

temp_df = pd.DataFrame(temperatures, index=cities, columns=months)
```

*Data Analysis*

```python
annual_avg_temp = temp_df.mean(axis=1)
highest_avg_temp_city = annual_avg_temp.idxmax()
lowest_avg_temp_city = annual_avg_temp.idxmin()
```

*Data Visualisation*

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for city in cities:
    if city in [highest_avg_temp_city, lowest_avg_temp_city]:
        plt.plot(months, temp_df.loc[city], label=city, linewidth=2.5, linestyle='--')
    else:
        plt.plot(months, temp_df.loc[city], label=city, alpha=0.7)

plt.title("Monthly Temperature Trends for Each City")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.heatmap(temp_df, cmap="coolwarm", annot=True, fmt=".1f")

plt.xlabel("Month")
plt.ylabel("City")
plt.title("Monthly Average Temperatures (C)")

plt.show()
```

Key Points about **Alpha** ():
Type I Error: This is the error of rejecting a true null hypothesis. Alpha is the probability of committing this error. For example, if  = 0.05, there is a 5% risk of rejecting the null hypothesis when it is actually true.

Common Values: The most commonly used alpha level is 0.05, but other values like 0.01 or 0.10 can also be used depending on the context and the desired level of confidence. A lower alpha level means a stricter criterion for rejecting the null hypothesis, reducing the risk of a Type I error but increasing the risk of a Type II error (failing to reject a false null hypothesis).

Decision Rule: In hypothesis testing, you compare the p-value (the probability of observing the test results under the null hypothesis) to the alpha level:

If the p-value  , you reject the null hypothesis.
If the p-value > , you fail to reject the null hypothesis.
Interpretation: Choosing an alpha level is a balance between being too lenient (risking more Type I errors) and being too strict (risking more Type II errors). The choice of alpha should reflect the consequences of making these errors in the specific context of the test.

Contextual Considerations: In fields where the cost of a Type I error is high (e.g., medical trials), a lower alpha (e.g., 0.01) might be chosen to minimize the risk of incorrectly concluding that a treatment is effective. Conversely, in exploratory research, a higher alpha might be acceptable.

**Skewness**
Definition: Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates whether the data points are skewed to the left or right of the mean.

Positive Skew (Right Skew): When a distribution has a long tail on the right side, it is said to be positively skewed. This means that the right tail (higher values) is longer or fatter than the left tail. In such distributions, the mean is typically greater than the median.

Negative Skew (Left Skew): When a distribution has a long tail on the left side, it is negatively skewed. This means that the left tail (lower values) is longer or fatter than the right tail. In these cases, the mean is typically less than the median.

Zero Skewness: A skewness of zero indicates a perfectly symmetrical distribution, such as a normal distribution.

**Kurtosis**
Definition: Kurtosis is a measure of the "tailedness" or the sharpness of the peak of the probability distribution. It provides insight into the extremity of deviations (outliers) from the mean.

High Kurtosis (Leptokurtic): Distributions with high kurtosis have heavy tails and a sharp peak. This indicates that more of the variance is due to infrequent extreme deviations (outliers). Such distributions have a higher likelihood of producing outliers.

Low Kurtosis (Platykurtic): Distributions with low kurtosis have light tails and a flatter peak. This suggests that the data are more evenly distributed around the mean, with fewer extreme deviations.

Normal Kurtosis (Mesokurtic): A normal distribution has a kurtosis of 3 (excess kurtosis of 0). This is considered the baseline for comparison.

Importance
Skewness helps identify the direction and degree of asymmetry in the data, which can affect statistical analyses and interpretations. For example, many statistical tests assume normality, and skewness can indicate deviations from this assumption.

Kurtosis provides information about the tails of the distribution, which is important for understanding the likelihood of extreme events. In finance, for example, high kurtosis might indicate a higher risk of extreme returns.

```python
from scipy.stats import skew, kurtosis

# Generate a random dataset of 1000 numbers from a normal distribution
random_data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate mean
mean_value = np.mean(random_data)
print("Mean:", mean_value)

# Calculate median
median_value = np.median(random_data)
print("Median:", median_value)

# Calculate variance
variance_value = np.var(random_data)
print("Variance:", variance_value)

# Calculate skewness
skewness_value = skew(random_data)
print("Skewness:", skewness_value)

# Calculate kurtosis
kurtosis_value = kurtosis(random_data)
print("Kurtosis:", kurtosis_value)
```

*Data creation*

```python
import numpy as np
import pandas as pd

# Create date range for January 2020
dates = np.arange('2020-01-01', '2020-02-01', dtype='datetime64[D]')
#date_range = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')

print("January 2020 Dates:\n", dates)

# Simulate daily temperature and humidity data
temperature_data = np.random.uniform(low=10, high=35, size=len(dates))
humidity_data = np.random.uniform(low=30, high=90, size=len(dates))

# Create a DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature_data,
    'Humidity': humidity_data
})
```

*Data Analysis*

```python
# Calculate mean, median, and standard deviation
temperature_mean = df['Temperature'].mean()
temperature_median = df['Temperature'].median()
temperature_std = df['Temperature'].std()

humidity_mean = df['Humidity'].mean()
humidity_median = df['Humidity'].median()
humidity_std = df['Humidity'].std()

print(f"Temperature - Mean: {temperature_mean}, Median: {temperature_median}, Std: {temperature_std}")
print(f"Humidity - Mean: {humidity_mean}, Median: {humidity_median}, Std: {humidity_std}")

# Identify dates with highest temperature and humidity
max_temp_date = df.loc[df['Temperature'].idxmax(), 'Date']
max_humidity_date = df.loc[df['Humidity'].idxmax(), 'Date']

print(f"Date with Highest Temperature: {max_temp_date}")
print(f"Date with Highest Humidity: {max_humidity_date}")
```

*Plotting Temperature and Humidity trends*

```python
import matplotlib.pyplot as plt

# Plot temperature and humidity trends
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Temperature'], label='Temperature (C)', color='red')
plt.plot(df['Date'], df['Humidity'], label='Humidity (%)', color='blue')
plt.title('Temperature and Humidity Trends in January 2020')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

*Window Function*

```python
# Calculate 7-day moving average
df['Temperature_MA'] = df['Temperature'].rolling(window=7).mean()
df['Humidity_MA'] = df['Humidity'].rolling(window=7).mean()

print("Data with Moving Averages:\n", df.head(10))
```

```python
# Plot original data and moving averages
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Temperature'], label='Original Temperature', color='red', alpha=0.5)
plt.plot(df['Date'], df['Temperature_MA'], label='7-Day MA Temperature', color='darkred')
plt.plot(df['Date'], df['Humidity'], label='Original Humidity', color='blue', alpha=0.5)
plt.plot(df['Date'], df['Humidity_MA'], label='7-Day MA Humidity', color='darkblue')
plt.title('Temperature and Humidity with 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
