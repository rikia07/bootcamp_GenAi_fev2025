Daily Challenge: NumPy, Pandas, and Matplotlib Integration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

générer un ensemble de données synthétiques représentant les températures mensuelles moyennes (en degrés Celsius) sur 12 mois dans 10 villes différentes. Les températures doivent être comprises entre -5 et 35 degrés.

```python
city = 10
month = 12
data_temp = np.random.uniform(-5, 35, size =(city, month))
print(data_temp)
```

Convertissez ce tableau NumPy en un Pandas DataFrame, en ajoutant les noms de villes comme index et les mois comme colonnes.

```python
months =  ["Janvier", "Février", "Mars", "Avril", "Mai", "Juin", 
        "Juillet", "Août", "Septembre", "Octobre", "Novembre", "Décembre"]
cities = ["Paris", "Londres", "Berlin", "Madrid", "Rome", 
          "New York", "Tokyo", "Pékin", "Sydney", "Moscou"]
df_temp = pd.DataFrame(data_temp, index= cities, columns= months )
```

```python
df_temp.head(12)
```

```python
average_year = df_temp.mean(axis=1)

print(average_year)
```

ville avec la temperature moyenne la plus élevée

```python
city_highest= average_year.idxmax()
print(f" the city with the hightest temperature is : ", city_highest )
```

temperature la plus basse par ville

```python
city_lowest= average_year.idxmin()
print(f" the city with the lowest temperature is : ", city_lowest )
```

visualisation

visualisation des températures moyenne

```python

plt.bar(cities, average_year)
plt.title('temperature by cities')
plt.xticks(rotation=90)
plt.xlabel('cities')
plt.ylabel('average temperature by year')
plt.show()
```
