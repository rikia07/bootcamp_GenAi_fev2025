**Analyse de Corrélations et Tests Statistiques**

T_test avec monthlyincome

```python
from scipy.stats import ttest_ind

# Séparer les groupes
attrition_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome']
attrition_no = df[df['Attrition'] == 'No']['MonthlyIncome']

# Effectuer le test T
stat, p_value = ttest_ind(attrition_yes, attrition_no)
print(f"Statistique du test T : {stat}, p-value : {p_value}")
```

** conclusion générale avec des recommandations startégiques **

Employee turnover poses a significant challenge for organizations, resulting in increased costs and the loss of valuable talent. Through an analysis of IBM’s HR data, we identified several **key factors contributing to employee attrition**:

Low salaries drive employees to explore more lucrative opportunities.

Long commute times cause stress and dissatisfaction.

Excessive overtime leads to burnout and fatigue.

Limited career growth and promotional opportunities prompt employees to leave.

Poor work-life balance raises the likelihood of resignation.

New hires often face adjustment difficulties, leading to higher turnover.

High-performing employees are frequently targeted by competitors offering better positions.

**Recommended Solutions to Reduce Attrition:**

Increase salaries for employees earning below market rates.

Provide flexible work arrangements for those with long commutes.

Limit overtime to improve overall work-life balance.

Strengthen training and development programs, along with promotion pathways, to retain top talent.
