Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
data = {'Age': [18, 21, 22, 19, 24, 30, 28, 35, 40, 45],
        'Panic Attacks?': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No']}
df = pd.DataFrame(data)

# Conversion des r√©ponses en valeurs num√©riques
df['Panic_Attack'] = df['Panic Attacks?'].map({'Yes': 1, 'No': 0})

sns.scatterplot(x=df['Age'], y=df['Panic_Attack'], color='red', marker='o', s=100)
plt.xlabel('Age')
plt.ylabel('Panic Attacks (0 = No, 1 = Yes)')
plt.title('Relation entre l\'√¢ge et les crises de panique')
plt.grid(True)

# Affichage du graphique
plt.show()
```
